```python
from rfdetr import RFDETRSmall
import torch
import numpy
print(torch.cuda.is_available())
print(torch.__version__)
print(numpy.__version__)
```

```python
model = RFDETRSmall() 
print(type(model))
```

```python
from pathlib import Path
from rfdetr.datasets.aug_config import AUG_CONSERVATIVE, AUG_AGGRESSIVE, AUG_AERIAL, AUG_INDUSTRIAL


DATASET_RFDETR = Path("C:\\Users\\chris\\Documents\\GitHub\\DINO-LungDet\\data\\processed")   # anpassen
OUTPUT_DIR = Path("./outputs/rfdetr_small_exp1")

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

results = model.train(
    dataset_dir=str(DATASET_RFDETR),
    output_dir=str(OUTPUT_DIR),
    epochs=200,
    batch_size=4,
    grad_accum_steps=1,
    lr=1e-4,
    device="cuda",
    early_stopping=True,
    early_stopping_patience=15,  # Wait 15 epochs before stopping
    early_stopping_min_delta=0.005,  # Require 0.5% mAP improvement
    early_stopping_use_ema=True,  # Track EMA model performance
    aug_config=AUG_CONSERVATIVE,
    print_freq=1,
    progress_bar=True
)
```
